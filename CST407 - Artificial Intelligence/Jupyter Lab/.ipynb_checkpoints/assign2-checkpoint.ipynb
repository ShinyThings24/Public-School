```python
#############################################
# Breaks down survival rates based on fare paid
#51 bins are used, with a range of 0-512.3. Each bin represents approx 10 dollars.
plt.title("Survival based on fare", fontsize = 20)
plt.hist(
    [titanic_train["Fare"].where(titanic_train["Survived"] == 0),
     titanic_train["Fare"].where(titanic_train["Survived"] == 1)],
     label=["Died", "Survived"], stacked=True, bins=51)
plt.legend()
plt.show()
```

```python
# pre-proccess data
titanic_train = pd.read_csv(datapath + "train.csv", thousands=',')
titanic_test = pd.read_csv(datapath + "test.csv", thousands=',')
# set index
titanic_train.set_index("PassengerId", inplace=True)
titanic_test.set_index("PassengerId", inplace=True)

# extract survived as training label
titanic_train_y = titanic_train["Survived"]
titanic_train_x = titanic_train.drop("Survived", axis=1)

# drop Name, SibSp, Parch, Ticket, Cabin
titanic_train_x.drop(["Name","SibSp","Parch","Ticket","Cabin"], axis=1, inplace=True)
titanic_test_x = titanic_test.drop(["Name","SibSp","Parch","Ticket","Cabin"], axis=1)
```

```python
# Standardize Data
# set up and run pre-processing pipeline
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer 

# pipeline for numeric attributes
num_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler())
    ])

#pipeline for catagorical attributes
cat_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(sparse=False)),
    ])

# combine pipelines to cover all attribute types present
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, ["Age", "Fare"]),
        ("cat", OneHotEncoder(), ["Pclass", "Sex", "Embarked"]),
    ])

# fit the completed pipeline to the training data
full_pipeline.fit(titanic_train_x)

# transform the training and testing data into standardized format
titanic_train_x = full_pipeline.transform(titanic_train_x)
titanic_test_x = full_pipeline.transform(titanic_test_x)
```

```python
# train model
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(random_state=42)
sgd.fit(titanic_train_x, titanic_train_y)
```

```python
# accuracy
from sklearn.model_selection import cross_val_score
cross_val_score(sgd, titanic_train_x, titanic_train_y, cv=3, scoring="accuracy")
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

titanic_train_pred = sgd.predict(titanic_train_x)
confusion_matrix(titanic_train_y, titanic_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score
print(precision_score(titanic_train_y, titanic_train_pred))
print(recall_score(titanic_train_y, titanic_train_pred))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(titanic_train_y, titanic_train_pred)
```

My precision and recall are inverted from the results the class demo got

```python
# try random forest classifier
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(random_state=42)
forest.fit(titanic_train_x, titanic_train_y)
print("Cross Validation:\n", cross_val_score(forest, titanic_train_x, titanic_train_y, cv=3, scoring="accuracy"))

titanic_train_pred = forest.predict(titanic_train_x)
print("Confusion Matrix:\n",confusion_matrix(titanic_train_y, titanic_train_pred))
print("Precision Score: ",precision_score(titanic_train_y, titanic_train_pred))
print("Recall Score: ",recall_score(titanic_train_y, titanic_train_pred))
print("ROC AUC: ", roc_auc_score(titanic_train_y, titanic_train_pred))
```

```python
# predict test data
titanic_test_pred = forest.predict(titanic_test_x)
output = pd.DataFrame()
output['PassengerId'] = titanic_test.index
output['Survived'] = titanic_test_pred
output.to_csv('submission.csv', index = False)
```
