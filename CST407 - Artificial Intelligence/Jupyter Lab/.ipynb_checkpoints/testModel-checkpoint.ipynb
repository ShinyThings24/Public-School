```python
import tensorflow as tf
from tensorflow import keras
```

```python
batch_size = 32
image_height = 512
image_width = 512
```

```python
import os
import numpy as np
# Establish datapath #
data_path = os.path.join("Datasets", "CanBottle", "")
model_path = os.path.join("Models", "CanBottle", "")
```

```python
# The names and number of classes we are identifying
class_names=["SmallBottle","LargeBottle","Can","Reject"]
class_count = 4
```

```python
# Load training data into datasets #
rand_seed = 123
test_split = 0.25
train = tf.keras.utils.image_dataset_from_directory(
    data_path,
    class_names=class_names,
    seed=rand_seed,
    batch_size=batch_size,
    image_size=(image_height, image_width),
    validation_split=test_split,
    subset="training")
# Load training data into datasets #
test = tf.keras.utils.image_dataset_from_directory(
    data_path,
    class_names=class_names,
    seed=rand_seed,
    batch_size=batch_size,
    image_size=(image_height, image_width),
    validation_split=test_split,
    subset="validation")
```

```python
model_save_path = model_path +'reducedModelR1.hdf5'
try:
    # Attempt to load pretrained model #
    model = keras.models.load_model(model_save_path)
except: 
    # Attempt to load pretrained model #
    # If not able to be loaded, generate and train the model #
    model = tf.keras.Sequential()
    # normalization layer, rescales pixel data between 0-1 #
    model.add(tf.keras.layers.Rescaling(1./255))

    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    

    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    


    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    

    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(128, activation='relu'))
    model.add(tf.keras.layers.Dense(class_count))

    model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'])

    with tf.device('/gpu:0'):
        model.fit(
            train,
            validation_data=test,
            epochs=5
        )
    model.save(model_save_path)
```

```python
# Load training data into datasets #
rand_seed = 24
test_split = 0.4
train = tf.keras.utils.image_dataset_from_directory(
    data_path,
    class_names=class_names,
    seed=rand_seed,
    batch_size=batch_size,
    image_size=(image_height, image_width),
    validation_split=test_split,
    subset="training")
# Load training data into datasets #
test = tf.keras.utils.image_dataset_from_directory(
    data_path,
    class_names=class_names,
    seed=rand_seed,
    batch_size=batch_size,
    image_size=(image_height, image_width),
    validation_split=test_split,
    subset="validation")
```

```python
model_save_path = model_path +'reducedModelSoftmax.hdf5'
try:
    # Attempt to load pretrained model #
    model = keras.models.load_model(model_save_path)
except: 
    # Attempt to load pretrained model #
    # If not able to be loaded, generate and train the model #
    model = tf.keras.Sequential()
    # normalization layer, rescales pixel data between 0-1 #
    model.add(tf.keras.layers.Rescaling(1./255))

    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    

    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    


    # Convolutional Layer, 32 filters, kernal size 3, 
    model.add(tf.keras.layers.Conv2D(32, 3, activation='relu'))
    # Pooling Layer #
    model.add(tf.keras.layers.MaxPooling2D())    

    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(128, activation='relu'))
    model.add(tf.keras.layers.Dense(class_count))
    model.add(tf.keras.layers.Dense(class_count, activation='softmax'))

    model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy'])
    with tf.device('/gpu:0'):
        model.fit(
            train,
            validation_data=test,
            epochs=5
        )
    model.save(model_save_path)
```

```python
from matplotlib import pyplot as plt
visualization_test = test.take(1)
```

```python
predictions = model.predict(visualization_test)
```
